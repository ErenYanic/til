### Problem 3: Reshape Matrix
**Source:** [Deep-ML](https://www.deep-ml.com/problems/3)

**Goal:** Write a Python function that reshapes a given matrix into a specified shape. if it cant be reshaped return back an empty list [ ]

```python
import numpy as np

def reshape_matrix(a: list[list[int|float]], new_shape: tuple[int, int]) -> list[list[int|float]]:
	#Write your code here and return a python list after reshaping by using numpy's tolist() method
    # Convert the input list to a numpy array
    matrix = np.array(a)
    
    # Get the total number of elements in the original matrix
    total_elements = matrix.size
    
    # Calculate the required number of elements for the new shape
    required_elements = new_shape[0] * new_shape[1]
    
    # Check if reshape is possible (total elements must match)
    if total_elements != required_elements:
        return []
    
    try:
        # Reshape the matrix using numpy
        reshaped_matrix = matrix.reshape(new_shape)
        
        # Convert back to a Python list and return
        return reshaped_matrix.tolist()
    
    except ValueError:
        # If reshape fails for any reason, return empty list
        return []
```

```python
print(reshape_matrix([[1,2,3,4],[5,6,7,8]], (4, 2)))    # Expected output: [[1, 2], [3, 4], [5, 6], [7, 8]]
print(reshape_matrix([[1, 2, 3, 4], [5, 6, 7, 8]], (1, 4)))    # Expected output: [] (empty list)
print(reshape_matrix([[1,2,3],[4,5,6]], (3, 2)))    # Expected output: [[1, 2], [3, 4], [5, 6]]
print(reshape_matrix([[1,2,3,4],[5,6,7,8]], (2, 4)))    # Expected output: [[1, 2, 3, 4], [5, 6, 7, 8]]
```
