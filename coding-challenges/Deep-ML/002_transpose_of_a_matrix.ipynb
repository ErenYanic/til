### Problem 2: Transpose of a Matrix
**Source:** [Deep-ML](https://www.deep-ml.com/problems/2)

**Goal:** Write a function that takes the transpose of a two-dimensional matrix

```python
def transpose_matrix(a: list[list[int|float]]) -> list[list[int|float]]:
    """
    Transpose a 2D matrix by swapping rows and columns.
    
    Args:
        a: A 2D matrix of shape (m, n)
    
    Returns:
        The transposed matrix of shape (n, m)
    """
    # Your code here
    return [list(row) for row in zip(*a)]
```

```python
print(transpose_matrix([[1, 2], [3, 4], [5, 6]]))   # Expected output: [[1, 3, 5], [2, 4, 6]]
print(transpose_matrix([[1, 2, 3], [4, 5, 6]]))  # Expected output: [[1, 4], [2, 5], [3, 6]]
print(transpose_matrix([[1, 2], [3, 4]]))   # Expected output: [[1, 3], [2, 4]]
print(transpose_matrix([[1, 2, 3]]))    # Expected output: [[1], [2], [3]]
print(transpose_matrix([[1], [2], [3]]))    # Expected output: [[1, 2, 3]]
```
